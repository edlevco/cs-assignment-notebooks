```python
# Final working solution
# I prompted ChatGPT with "Make this code the best and most efficent queue solution"


from collections import deque

def countStudents(self, students: List[int], sandwiches: List[int]) -> int:
    students = deque(students)
    sandwiches = deque(sandwiches)

    count_rotations = 0

    while students and count_rotations < len(students):
        if students[0] == sandwiches[0]:
            students.popleft()
            sandwiches.popleft()
            count_rotations = 0
        else:
            students.append(students.popleft())
            count_rotations += 1

    return len(students)

```

```python
# Final working solution
# Used ChatGPT with prompt "Can you fix this code and make it the most efficient binary search"

class Solution:
    def firstBadVersion(self, n: int) -> int:
        left, right = 1, n

        while left < right:
            mid = left + (right - left) // 2  # prevents overflow

            if isBadVersion(mid):
                right = mid      # mid might be the first bad
            else:
                left = mid + 1   # mid is good ‚Üí eliminate left half

        return left  # left == right ‚Üí first bad version
```

This solution uses the Dutch national flag algorithm to solve this problem

This algorithm starts by placing both low and mid pointers at the start of the array and high at the end

If array[mid] is 0, swap it with low and increase mid and low by 1

If array[mid] is 1, swap it with mid and only increase mid by 1

if array[mid] is 2, swap it with high and decrease high by 1

```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:

        low, mid, high = 0, 0, len(nums) - 1

        while mid < high:
            if nums[mid] == 0:
                nums[mid], nums[low], nums[low], nums[mid]
                low +=1
                mid +=1
            if nums[mid] == 1:
                mid += 1
            else:
                nums[mid], nums[high] = nums[high], nums[mid]
                high -=1
```
